```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import csv
```

```python
driver = webdriver.Chrome(executable_path=r"E:\adobe\CDE\chromedriver-win64\chromedriver-win64\chromedriver.exe")
driver.get('https://www.psx.com.pk/market-summary/')
```

```python
tables = driver.find_elements(By.CLASS_NAME,'table-responsive')
titles = driver.find_elements(By.TAG_NAME, 'th')
titles_texts = [title.text.strip() for title in titles]
print(titles_texts)
```

```python
table_data = []
for table in tables:
    try:
        rows = table.find_elements(By.TAG_NAME, 'tr')
        if not rows:
            continue  # Skip empty tables
        
        # Extract headers from the first row of the table
        header_row = rows[0]
        headers = [th.text.strip() for th in header_row.find_elements(By.TAG_NAME, 'th')]
        table_data.append(headers)  # Add headers once per table
        
        # Process data rows (skip the header row)
        for row in rows[1:]:
            cols = row.find_elements(By.TAG_NAME, 'td')
            if len(cols) == 8:  # Validate column count
                row_data = [col.text.strip() for col in cols]
                table_data.append(row_data)
    except Exception as e:
        print(f"Error processing table: {e}")
```

```python
table_data
```

```python
header = ['SCRIP', 'LDCP', 'OPEN', 'HIGH', 'LOW', 'CURRENT', 'CHANGE', 'VOLUME']

with open('psx_data.csv', 'w', newline='', encoding='utf-8-sig') as f:
    writer = csv.writer(f)
    writer.writerow(header)      # Write headers
    writer.writerows(table_data)  # Write all rows

print("Data successfully saved to 'psx_data.csv'")
```
